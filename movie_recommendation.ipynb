```python
ratings_with_movies = ratings.merge(movies, on='movie_id')
```

```python
ratings_with_movies.head()
```

```python
ratings_with_movies.shape
```

```python
 num_ratings = ratings_with_movies.groupby('movies')['rating'].count().reset_index()
```

```python
num_ratings.head(25)
```

```python
num_ratings.rename(columns={'rating': 'num_of_ratings'}, inplace=True)
```

```python
final_ratings=ratings_with_movies.merge(num_ratings, on='movies')
```

```python
final_ratings.head()
```

```python
final_ratings.shape
```

```python
final_ratings = final_ratings[final_ratings['num_of_ratings'] >= 50]
```

```python
final_ratings.head()
```

```python
final_ratings.shape
```

```python
final_ratings.duplicated().sum()
```

```python
final_ratings.drop_duplicates(inplace=True)
```

```python
final_ratings.duplicated().sum()
```

```python
final_ratings.shape
```

```python
movies_pivot = final_ratings.pivot_table(columns='user_id', index='movies', values='rating')
```

```python
movies_pivot.head()
```

```python
movies_pivot.shape
```

```python
movies_pivot.fillna(0, inplace=True)
```

```python
from scipy.sparse import csr_matrix

movies_sparse = csr_matrix(movies_pivot)
```

```python
movies_sparse
```

```python
from sklearn.neighbors import NearestNeighbors

model = NearestNeighbors(algorithm='brute')
```

```python
model.fit(movies_sparse)
```

```python
distances, suggestions = model.kneighbors(movies_pivot.iloc[237, :].values.reshape(1, -1), n_neighbors=6)
```

```python
distances
```

```python
suggestions
```

```python
for i in range(len(suggestions)):
    print(movies_pivot.index[suggestions[i]])
```

```python
import numpy as np
def recommend(movie_name):
    movie_id = np.where(movies_pivot.index == movie_name)[0][0]
    distances, suggestions = model.kneighbors(movies_pivot.iloc[movie_id, :].values.reshape(1, -1), n_neighbors=6)

    for i in range(len(suggestions)):
      movies = movies_pivot.index[suggestions[i]]
      for j in movies:
        print(j)
```

```python
movie_name = 'Bats (1999)'
recommend(movie_name)
```
